# 프레젠테이션 준비

### 블러 이미지 생성

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = "Datasets/Val/Child/0969_1980_08_00000048_F.png"

# 이미지 읽기 (BGR→RGB 변환)
img = cv2.imread(img_path)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# 블러 처리 (커널 사이즈: 15x15)
blurred_img = cv2.blur(img, (15, 15))

# Matplotlib으로 이미지 시각화
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].imshow(img)
axes[0].set_title("Original Image")
axes[0].axis('off')

axes[1].imshow(blurred_img)
axes[1].set_title("Blurred Image (15x15)")
axes[1].axis('off')

plt.tight_layout()
plt.show()
```

### 다양한 블러 방식 적용

```python
img_path = "Datasets/Val/Child/0969_1980_08_00000048_F.png"

# 원본 이미지 읽기 (RGB)
img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)

# Average Blur
avg_blur = cv2.blur(img, (15, 15))

# Gaussian Blur
gauss_blur = cv2.GaussianBlur(img, (15, 15), 0)

# Median Blur
median_blur = cv2.medianBlur(img, 15)

# Bilateral Filter
bilateral_blur = cv2.bilateralFilter(img, d=15, sigmaColor=75, sigmaSpace=75)

# Motion Blur 커널 생성
kernel_size = 15
motion_kernel = np.zeros((kernel_size, kernel_size))
motion_kernel[int((kernel_size-1)/2), :] = np.ones(kernel_size)
motion_kernel /= kernel_size

motion_blur = cv2.filter2D(img, -1, motion_kernel)

# 이미지 시각화
blurred_images = [avg_blur, gauss_blur, median_blur, bilateral_blur, motion_blur]
titles = ["Average Blur", "Gaussian Blur", "Median Blur", "Bilateral Filter", "Motion Blur"]

fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.ravel()

# 원본 이미지 표시
axes[0].imshow(img)
axes[0].set_title("Original Image")
axes[0].axis('off')

# 각 블러 처리된 이미지 표시
for i, (blurred, title) in enumerate(zip(blurred_images, titles), start=1):
    axes[i].imshow(blurred)
    axes[i].set_title(title)
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 로그 파일 경로
log_path = "logs/simple_20250711_223739.csv"

# 로그 파일 읽기
df = pd.read_csv(log_path)  # 열: epoch, loss, train_acc, val_acc

# 시각화
plt.figure(figsize=(10, 5))
plt.plot(df['epoch'], df['train_acc'], label='Train Accuracy (%)', marker='o')
plt.plot(df['epoch'], df['val_acc'], label='Validation Accuracy (%)', marker='s')
plt.xlabel('Epoch')
plt.ylabel('Accuracy (%)')
plt.title('Train vs Validation Accuracy (SimpleCNN)')
plt.legend()
plt.xticks(df['epoch'])
plt.grid(True)
plt.tight_layout()
plt.show()
```
